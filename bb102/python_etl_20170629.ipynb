```python
# Classes
class Human:
    
    # class attr
    wiki_name = "H. sapiens"
    
    def __init__(self, name):
        self.name = name
    
    def say(self, msg):
        print("%s: %s"%(self.name, msg))
    
    @classmethod
    def get_species(cls):
        return cls.wiki_name
```

```python
ian = Human("Ian")
```

```python
ian.say("Hello word")
```

```python
ian.wiki_name
```

```python
ian.get_species()
```

```python
class Batman(Human):
    def __init__(self, name, superpower):
        super(Batman, self).__init__(name)
        self.superpower = superpower
        
    def say(self, msg):
        print("Bat man is saying..")
        super(Batman, self).say(msg)
    
    def punch():
        print("Boom")
```

```python
bruce = Batman("Bruce Wayne", "fly")
```

```python
bruce.say("Hi I am batman")
```
