```python
# List comprehension
USD_price = []
TWD_price = [100, 88, 33]

for price in TWD_price:
    tmp_price = price * 30
    print(tmp_price)
    USD_price.append(tmp_price)
USD_price
```

```python
USD_price = [price * 30 for price in TWD_price]
USD_price
```

```python
# higher order functions
list(map(lambda x: x * 30, TWD_price))
```

```python
# Jieba
jieba.set_dictionary('./dict.txt.big')
```

```python
with open('./cai_speech.txt') as f:
    speech = f.read()
```

```python
wd_count = Counter()
for word in jieba.cut(speech):
    print(word)
    if word in wd_count:
        wd_count[word] += 1
    else:
        wd_count[word] = 1
```

```python
wd_count.most_common(20)
```

```python
# Classes
class Human:
    
    # class attr
    wiki_name = "H. sapiens"
    
    def __init__(self, name):
        self.name = name
    
    def say(self, msg):
        print("%s: %s"%(self.name, msg))
    
    @classmethod
    def get_species(cls):
        return cls.wiki_name
```

```python
ian = Human("Ian")
```

```python
ian.say("Hello word")
```

```python
ian.wiki_name
```

```python
ian.get_species()
```

```python
class Batman(Human):
    def __init__(self, name, superpower):
        Human.__init__(self, name)
        self.superpower = superpower
    
    def punch():
        print("Boom")
```

```python
bruce = Batman("Bruce Wayne", "fly")
```

```python
bruce.say("Hi I am batman")
```
